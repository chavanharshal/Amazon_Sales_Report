# Data Cleaning

```python
# Fill missing values in 'currency' with 'INR'
data['currency'].fillna('INR', inplace=True)
```

```python
# Fill missing values in 'Amount' with the column mean
data['Amount'].fillna(data['Amount'].mean(), inplace=True)
```

```python
data.isnull().sum()
```

```python
# Fill missing values in 'ship-city' and 'ship-state' with the mode
data['ship-city'].fillna(data['ship-city'].mode()[0], inplace=True)
data['ship-state'].fillna(data['ship-state'].mode()[0], inplace=True)
```

```python
# Fill missing values in 'ship-city' and 'ship-state' with the mode
data['ship-postal-code'].fillna(data['ship-postal-code'].mode()[0], inplace=True)
data['ship-country'].fillna(data['ship-country'].mode()[0], inplace=True)
```

```python
data.isnull().sum()
```

```python
data['fulfilled-by'].fillna('Easy Ship', inplace=True)
```

```python
data.isnull().sum()
```

```python
data.drop(columns=['New', 'PendingS'], inplace=True)
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace = True)
```

```python
data.duplicated().sum()
```

```python
# Save the updated DataFrame to a CSV file
data.to_csv('Updated_Amazon_Sales_Report.csv', index=False)
```

```python
# Ensure 'Date' column is in datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Resample by month to analyze monthly trends
sales_over_time = data.resample('M', on='Date').agg(
    Total_Revenue=('Amount', 'sum'),
    Total_Orders=('Order ID', 'count'),
    Average_Order_Value=('Amount', 'mean')
).reset_index()

# Display the aggregated data
print(sales_over_time.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the Seaborn style
sns.set_theme(style="whitegrid")

# Initialize the figure
plt.figure(figsize=(14, 7))

# Plot Total Revenue
sns.lineplot(data=sales_over_time, x='Date', y='Total_Revenue', label='Total Revenue', color='blue')

# Add chart details
plt.title('Monthly Sales Trends', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Revenue (in currency)', fontsize=12)
plt.legend(title='Metrics')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Calculate total revenue and quantity for each category
category_totals = data.groupby('Category')[['Amount', 'Qty']].sum()

# Sort by Total Revenue in descending order
category_analysis = category_totals.sort_values('Amount', ascending=False)

# Rename columns for clarity
category_analysis.rename(columns={'Amount': 'Total_Revenue', 'Qty': 'Total_Quantity'}, inplace=True)

# Display top categories
print("Top Categories by Revenue:\n", category_analysis.head())
```

```python
# Calculate and sort sales by size
size_analysis = (
    data.groupby('Size')[['Amount', 'Qty']]
    .sum()
    .rename(columns={'Amount': 'Total_Revenue', 'Qty': 'Total_Quantity'})
    .sort_values('Total_Revenue', ascending=False)
)

# Display top sizes
print("Top Sizes by Revenue:\n", size_analysis.head())
```

```python
# Calculate total revenue and orders by fulfillment method
fulfillment_totals = data.groupby('Fulfilment').agg(
    Total_Revenue=('Amount', 'sum'),
    Total_Orders=('Order ID', 'count')
)

# Calculate cancellation rate by fulfillment method
cancellation_rate = (
    data[data['Status'] == 'Cancelled']
    .groupby('Fulfilment')['Status']
    .count()
    / data.groupby('Fulfilment')['Status'].count()
) * 100

# Combine results into a single DataFrame
fulfillment_analysis = fulfillment_totals.copy()
fulfillment_analysis['Cancellation_Rate'] = cancellation_rate

# Replace NaN values with 0 (in case there are fulfillment methods without cancellations)
fulfillment_analysis['Cancellation_Rate'] = fulfillment_analysis['Cancellation_Rate'].fillna(0)

# Display fulfillment performance
print("Fulfillment Analysis:\n", fulfillment_analysis)
```

```python
# Calculate customer metrics
customer_totals = data.groupby('Order ID')[['Amount']].sum()
customer_totals.rename(columns={'Amount': 'Total_Revenue'}, inplace=True)

# Count total orders per customer
customer_totals['Total_Orders'] = data.groupby('Order ID')['Order ID'].count()

# Sort customers by Total Revenue in descending order
customer_analysis = customer_totals.sort_values(by='Total_Revenue', ascending=False)
# Display top customers by revenue
print("Top Customers by Revenue:\n", customer_analysis.head())
```

```python
# Add percentage contribution to total revenue
total_revenue = customer_analysis['Total_Revenue'].sum()
customer_analysis['Revenue_Percentage'] = (customer_analysis['Total_Revenue'] / total_revenue) * 100
import matplotlib.pyplot as plt

# Plot the top 10 customers
top_customers = customer_analysis.head(10)
top_customers['Total_Revenue'].plot(kind='bar', figsize=(12, 6), color='blue', title='Top 10 Customers by Revenue')
plt.xlabel('Customer (Order ID)')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
# Aggregate revenue and orders by state
geo_revenue = data.groupby('ship-state')['Amount'].sum().rename('Total_Revenue')
geo_orders = data.groupby('ship-state')['Order ID'].count().rename('Total_Orders')

# Combine results into a single DataFrame
geo_analysis = pd.concat([geo_revenue, geo_orders], axis=1)

# Sort states by Total Revenue in descending order
geo_analysis = geo_analysis.sort_values(by='Total_Revenue', ascending=False)

# Display top states
print("Top States by Revenue:\n", geo_analysis.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Prepare top 10 states for visualization
top_states = geo_analysis.head(10).reset_index()

# Plot the data using Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(data=top_states, x='ship-state', y='Total_Revenue', palette='Blues_d')
plt.title("Top 10 States by Revenue", fontsize=16)
plt.xlabel("State", fontsize=12)
plt.ylabel("Total Revenue", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
